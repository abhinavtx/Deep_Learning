```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sonar_dataset.csv')
df.head()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
y = pd.get_dummies(y, drop_first=True)
y.sample(5) # R --> 1 and M --> 0
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)
```

```python
## Now using Deep Learning without --> Dropout Layer
import tensorflow as tf
from tensorflow import keras
```

```python
print("The shape of training and testing is: ",x_train.shape," ",x_test.shape)
```

```python
model = keras.Sequential([
    ## Here we we have given value as 60 as there are 60 columns, so 60 input in input layer
    keras.layers.Dense(x_train.shape[1],input_dim=60, activation = 'relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=100, batch_size=8)
## In this batch size = 8, means we are following batch gradient descent
```

```python
model.evaluate(x_test,y_test)
```

```python
yPred = model.predict(x_test)
## Its a 2d array so we have to reshape it
yPred = yPred.reshape(len(yPred),)
```

```python
yPred = np.round(yPred)
```

```python
modelD = keras.Sequential([
    ## Here we we have given value as 60 as there are 60 columns, so 60 input in input layer
    ## Here first 60 is number of neurons in input layer and input_dim = 60 means there are 60 number of features in input
    keras.layers.Dense(60,input_dim=60, activation = 'relu'),
    keras.layers.Dropout(0.4),
    ## This number how much portion of the neurons are to be dropped
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.4),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.4),
    keras.layers.Dense(1, activation='sigmoid')
])

modelD.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

modelD.fit(x_train, y_train, epochs=100, batch_size=8)
## In this batch size = 8, means we are following batch gradient descent
```

```python
## We apply dropout layer after every hidden layer
```

```python
modelD.evaluate(x_test,y_test)
```
